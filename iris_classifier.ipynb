```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("C:/Users/asus/Desktop/Iris.csv")
```

```python
dataset.head()
```

```python
dataset.shape   # gives (x,y) where x is num of rows and y num of cols
```

```python
dataset.Species.unique() # this tells us the diff number of classifiers of the dataset
```

```python
setosa_df=dataset.loc[dataset['Species']=='Iris-setosa']

versicolor_df=dataset.loc[dataset['Species']=='Iris-versicolor']

virginica_df=dataset.loc[dataset['Species']=='Iris-virginica']
```

#Performing univariate analysis using matplotlib

```python
plt.plot(setosa_df['PetalLengthCm'],np.zeros_like(setosa_df['PetalLengthCm']),'o') #plotting the feature petallength wit respect to first classifier. We will do this for every class
plt.plot(versicolor_df['PetalLengthCm'],np.zeros_like(versicolor_df['PetalLengthCm']),'o')  #zeros_like adds zeros since we are using only one axis, we will give zero for other axis
plt.plot(virginica_df['PetalLengthCm'],np.zeros_like(virginica_df['PetalLengthCm']),'o')

#adding 'o' makes the dots more larger
plt.xlabel('Petal Length in cm')
plt.show()
```

```python
# Performing bivariate analysis using seaborn
```

```python
sns.FacetGrid(dataset,hue="Species",size=5).map(plt.scatter,"PetalLengthCm","SepalWidthCm").add_legend();
plt.show()

#we have used 2 cols petallengthcm and sepalwidthcm, hue means the classifier
```
